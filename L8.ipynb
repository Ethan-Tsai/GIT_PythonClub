```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #畫圖用
import seaborn as sns #資料處理
```

# 相關矩陣

```python
plt.figure(figsize=(12, 8))
#heatmap: 相關矩陣畫圖
#sns.heatmap(資料, 預設值(不必管), 顏色(google cmap for more information), 寬度)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('visualization')
plt.show()
```

```python
# 散點圖可視化
plt.figure(figsize=(10, 6))
sns.scatterplot(x='longitude', y='latitude', hue='aqi', data=df, palette='viridis', size='aqi', sizes=(20, 200))
plt.title('AQI_scatter')
plt.show()
```

```python
# 繪製 AQI 與風速的關係
plt.figure(figsize=(10, 6))
sns.scatterplot(x='wind_speed', y='aqi', data=df)
plt.title('AQI_wind')
plt.show()
```

### 資料集:https://data.gov.tw/dataset/8116

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
 
df = pd.read_csv('歷年國內主要觀光遊憩據點遊客人數月別統計.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.value_counts('類型')
```

```python
df.value_counts('縣市別')
```

```python
df.groupby(['縣市別', '類型']).count()
#pd.set_option('display.max_rows', None) #顯示完整數據
```

```python
#設定新row
rows = (df['年別'] == 2019) & (df['縣市別'] == '臺北市')
columns = ['細分', '1月', '2月', '3月'] #選擇條件
result = df.loc[rows, columns].head(10) #原本.loc存進([rows, columns,jead) result
result.set_index('細分', inplace=True)
 
chart = result.plot(figsize=(10, 5))  # 圖表大小
 
font = FontProperties(fname=r'NotoSansTC-VariableFont_wght.ttf')
 
for label in chart.get_xticklabels():
    label.set_fontproperties(font)  # 設定x軸每一個細分(景點名稱)字型

plt.title('2019年臺北市各景點旅客人數', fontproperties=font)  #圖表標題
plt.xlabel('細分(景點名稱)', fontproperties=font)  #x軸說明文字
plt.ylabel('人數', fontproperties=font)  #y軸說明文字
plt.legend(prop=font)  #圖例
 
plt.show()
```
